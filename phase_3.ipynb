# Advance Operations with BUsiness expamles

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]

sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("===== Zomato sales analysis =====")
print("\nSales data Shape: ",sales_data.shape)  # first thing to do 
print("\n Sales data of first three restorents: \n",sales_data[0:3])
```

### Total sales per year

```python
yearly_total = np.sum(sales_data[:, 1:], axis=0)
# sales_data[:, 1:] → selects all rows, but from column index 1 onwards (2021–2024 sales only).
# we choese axis=0 beause we need to find out total zomato sales from every restorent each year
print("total sales from each restorent: ",yearly_total )
```

### Minimam & Maximam sales per restorent

```python
min_sales = np.min(sales_data[:,1:],axis=1)
# we choese axis=1 beause we need to find the minimam sales of every restorent from the data
print('Minimam sales of each restorent: ',min_sales)

max_sales = np.max(sales_data[:,1:],axis=1)
print('Maximam sales of each restorent: ',max_sales)
```

### Average sales per restorent

```python
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("Average sales of each restorent: ", avg_sales)
```

### monthly average per restorent per year

```python
monthly_avg = sales_data[:, 1:] / 12
print("Monthly average: \n", monthly_avg)
```

### Maximam & Minimam sales per year for Zomato 

```python
max_sales_per_year=np.max(sales_data[:,1:],axis=0)
print("Maximam sales for Zomato each year: ",max_sales_per_year)

min_sales_per_year=np.min(sales_data[:,1:],axis=0)
print("Minimam sales for Zomato each year: ",min_sales_per_year)
```

### Cumulative sales per restorent

```python
cumsu_m = np.cumsum(sales_data[:, 1:], axis=1)
print("Cumulative sales of each restorents: \n", cumsu_m)

plt.Figure(figsize=(10, 6))
plt.plot(np.mean(cumsu_m, axis=0))
plt.title("Average Cumulative sales for Zomato accross all restorents: ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
