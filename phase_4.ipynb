# saving and loading data

#### there are better librarys to do these tasks  its just for understanding 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr_1 = np.array([[1, 2, 3], [4, 5, 6]])
arr_2 = np.random.rand(3, 3)
arr_3 = np.zeros((4, 4))
```

#### saves arrays or data in a file

```python
# first argument is the name and path of the file and the second argument is the arr or data you want to store
np.save("array_1.npy",arr_1)  
np.save("array_2.npy",arr_2)
np.save("array_3.npy",arr_3)
```

#### load data from file

```python
# provide file path
loded_arr_1 = np.load("array_1.npy")
print(loded_arr_1)
```

### loding an image in the form of matrices & numbers of its pixcels

```python
try:
    logo = np.load("numpy-logo.npy")

    plt.Figure(figsize=(10, 5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("numpy-logo")
    plt.grid(False)

    dark_logo = 1-logo #inverting the color of each pixcel in the image
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("numpy-dark-logo")
    plt.grid(False)

except FileNotFoundError:
    print("numpy logo not found")
```
